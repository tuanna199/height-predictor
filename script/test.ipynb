```python
import json
import pandas as pd
```

```python
with open("data/Height_Wood_2014.json") as file:
    height_json = json.load(file)
```

```python
relevant_snps = height_json["relevantSNPs"]
relevant_snps
```

```python
snps_list = []

for dict in relevant_snps:
    snps_list.append(dict["snp"])
```

```python
len(snps_list)
```

```python
df = pd.DataFrame(snps_list, columns=["snp"])
```

```python
# Specify the file path
file_path = "data/height_snps.txt"
```

```python
# Write the DataFrame to a CSV file
df.to_csv(file_path, index=False, header=None)
```
